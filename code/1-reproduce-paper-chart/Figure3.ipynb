# Figure 3. The feature importance from DNA/RNA/Protein.

## ðŸ’« DNA

1. From figure 3 we knew the best sequence length is 111bp.
2. Under 111bp, for 5-fold validation, the feature importance are displayed as below.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load data
ROOT_PATH = "../../DNA_features/fig_5foldValidation/train8502_DNAfea_81bp_DART_selected_features_importance__"
df = pd.read_csv(f"{ROOT_PATH}fold1.csv", usecols=['Feature', 'importance'])
for i in range(2, 6):
    tdf = pd.read_csv(f"{ROOT_PATH}fold{i}.csv", usecols=['Feature', 'importance'])
    df = df.merge(tdf, how='left', left_on='Feature', right_on='Feature', suffixes=('', f'_{i}'))

# Calculate Mean and Std
df["mean"] = df.iloc[:, 1:6].mean(axis=1)
df["std"] = df.iloc[:, 1:6].std(axis=1)

# Calculate overall median
median_val = np.median(df.iloc[:, 1:6].values.flatten())

# Sort by Mean value
df_sorted = df.sort_values("mean", ascending=False)

# Assign color based on bigger than median or not
colors = df_sorted["mean"].apply(lambda x: "#0094ae" if x > median_val else "#dadada")

# Plot
fig, ax = plt.subplots(figsize=(6, 7))
y_pos = np.arange(len(df_sorted))

# Border setting of bars
bars = ax.barh(y_pos, df_sorted["mean"], xerr=df_sorted["std"], color=colors, capsize=5, edgecolor='none')

ax.set_yticks(y_pos)
ax.set_yticklabels(df_sorted["Feature"], fontfamily='DejaVu Sans')
ax.set_xlabel("DNA Importance (Â± Std)", fontfamily='DejaVu Sans', fontsize=14)

# Invert y axis
ax.invert_yaxis()

# Border setting of figure
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['bottom'].set_color('#666')
ax.spines['left'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

# Font setting
ax.tick_params(axis='x', labelsize=12, colors='#777')
ax.tick_params(axis='y', labelsize=12, colors='#777')

for label in ax.get_xticklabels():
    label.set_fontfamily('DejaVu Sans')
for label in ax.get_yticklabels():
    label.set_fontfamily('DejaVu Sans')

plt.tight_layout()
plt.savefig("./Figure3A.png", dpi=300)
plt.show()
```

## ðŸ’« Protein

1. From figure 3 we knew the best sequence length is 17aa.
2. Under 17aa, for 5-fold validation, the feature importance are displayed as below.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load data
ROOT_PATH = "../../Protein_features/5foldValidation_reproduce/train8502_proteinfea_17aa_DART_selected_features_importance__"
df = pd.read_csv(f"{ROOT_PATH}fold1.csv", usecols=['Feature', 'importance'])
for i in range(2, 6):
    tdf = pd.read_csv(f"{ROOT_PATH}fold{i}.csv", usecols=['Feature', 'importance'])
    df = df.merge(tdf, how='left', left_on='Feature', right_on='Feature', suffixes=('', f'_{i}'))

# Calculate Mean and Std
df["mean"] = df.iloc[:, 1:6].mean(axis=1)
df["std"] = df.iloc[:, 1:6].std(axis=1)

# Calculate overall median
median_val = np.median(df.iloc[:, 1:6].values.flatten())

# Sort by Mean value
df_sorted = df.sort_values("mean", ascending=False)

# Assign color based on bigger than median or not
colors = df_sorted["mean"].apply(lambda x: "#7bb34a" if x > median_val else "#dadada")

# Plot
fig, ax = plt.subplots(figsize=(6, 7))
y_pos = np.arange(len(df_sorted))

# Horizontal bars
bars = ax.barh(y_pos, df_sorted["mean"], xerr=df_sorted["std"], color=colors, capsize=5, edgecolor='none')

# Set y-axis ticks and labels
ax.set_yticks(y_pos)
ax.set_yticklabels(df_sorted["Feature"], fontfamily='DejaVu Sans')

# Axis label
ax.set_xlabel("Protein Importance (Â± Std)", fontfamily='DejaVu Sans', fontsize=14)
ax.xaxis.set_label_coords(0.5, -0.05)

# Flip both axes
ax.invert_yaxis()     # Highest importance at top
ax.invert_xaxis()     # Bars drawn from right to left

# Move y-axis to right
ax.yaxis.tick_right()
ax.yaxis.set_label_position('right')

# Border styling
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['bottom'].set_color('#666')
ax.spines['left'].set_visible(False)
ax.yaxis.set_ticks_position('right')
ax.xaxis.set_ticks_position('bottom')

# Fonts and colors
ax.tick_params(axis='x', labelsize=12, colors='#777')
ax.tick_params(axis='y', labelsize=12, colors='#777')

for label in ax.get_xticklabels():
    label.set_fontfamily('DejaVu Sans')
for label in ax.get_yticklabels():
    label.set_fontfamily('DejaVu Sans')

# Format x-axis ticks
ax.xaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))
ax.xaxis.set_label_coords(0.9, -0.05)  # 0.5æ°´å¹³å±…ä¸­ï¼Œ-0.12å¾€ä¸‹ç§»é¿å…è¢«yæ ‡ç­¾é®æŒ¡

plt.tight_layout()
plt.savefig("./Figure3C_flipped.png", dpi=300)
plt.show()
```
